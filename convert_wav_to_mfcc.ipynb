```python
def convert_wav_to_mfcc(save_path, dataset_path, duration_segment = 6, shift = 3,num_mfcc=13, n_fft=2048, hop_length=512, fmin = 0,start_at = 0):

    classnames = []

    info = {
        'filename': [],
        'genus': [],
        'species': [],
        'duration': [],
        'label': [],
    }

    mfccs = {
        "label": [],
        "mfcc": []
    }

    sample_rate = 22050
    samples_per_segment = duration_segment * sample_rate
    samples_per_shift = shift * sample_rate

    for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):
        if i == 0:
            for filename in tqdm(filenames[start_at::10],desc ="file: "):

                filepath = os.path.join(dirpath,filename)
                try:
                    signal, sr = librosa.load(filepath)
                except RuntimeError:
                    print('RuntimeError')
                    continue
                duration = librosa.get_duration(signal)
                print('duration: ',duration)
                filename_split = filename.split('-')
                genus = str(filename_split[0])
                species = str(filename_split[1])
                classname = genus+'-'+ species
                print(classname)
                label = int(classnames_info[classnames_info.classname == classname]['id'])
                print(label)
                num_segment = int(math.floor(duration - duration_segment)// shift)
                print('num_segment: ',num_segment)

                for n in range(num_segment):
                    start = samples_per_shift * n
                    finish = start + samples_per_segment

                    mfcc = librosa.feature.mfcc(signal[start:finish], sample_rate, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length ,fmin = 0)
                    mfcc = mfcc.T

                    mfccs["mfcc"].append(mfcc.tolist())
                    mfccs["label"].append(label)

    with open(save_path, "w") as fp:
        tqdm(json.dump(mfccs, fp, indent=4),desc ="mfccs saving ...")
    

    return mfccs
```
